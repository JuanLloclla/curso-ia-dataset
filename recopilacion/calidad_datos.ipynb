```python
import pandas as pd

df = pd.read_csv("../2020-Apr.csv")

print("\n=== VALORES NULOS ===")
print(df.isnull().sum())
print("\nPorcentaje de nulos:")
print((df.isnull().sum()/len(df) * 100).round(2))
```

```python
print("\n=== DUPLICADOS EXACTOS ===")
dupes = df.duplicated(subset=['event_time','event_type','product_id','user_id']).sum()
print(f"Duplicados: {dupes:,}")
```

```python
print("\n=== PRICE ===")
print(f"Min: {df['price'].min()}")
print(f"Max: {df['price'].max()}")
```
