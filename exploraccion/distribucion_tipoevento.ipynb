```python
import pandas as pd

df = pd.read_csv("../2020-Apr.csv")

# Distribución de event_type
print("\n=== DISTRIBUCIÓN EVENT_TYPE ===")
counts = df['event_type'].value_counts()
pct = df['event_type'].value_counts(normalize=True) * 100
print(pd.DataFrame({'Registros': counts, 'Porcentaje': pct.round(2)}))
```
